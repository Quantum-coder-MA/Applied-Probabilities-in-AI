```python
# =============================================================================
# FORMULE DE BAYES MANUELLE
# Exemple : Probabilité qu'une tumeur soit maligne si radius_mean > seuil
# =============================================================================

import numpy as np

# -------------------------------
# Paramètre choisi
# -------------------------------
seuil = 15.0  # seuil du radius_mean

# -------------------------------
# ÉTAPE 0 : Informations dataset
# -------------------------------
N_total = len(df)

N_M = (df['diagnosis'] == 'M').sum()  # Maligne
N_B = (df['diagnosis'] == 'B').sum()  # Bénigne


# -------------------------------
# ÉTAPE 1 : Probabilités a priori
# -------------------------------
P_M = N_M / N_total
P_B = N_B / N_total


# -------------------------------
# ÉTAPE 2 : Vraisemblances
# -------------------------------
P_R_M = (df[df['diagnosis'] == 'M']['radius_mean'] > seuil).mean()

P_R_B = (df[df['diagnosis'] == 'B']['radius_mean'] > seuil).mean()


# -------------------------------
# ÉTAPE 3 : Évidence
# -------------------------------
P_R = P_R_M * P_M + P_R_B * P_B


# -------------------------------
# ÉTAPE 4 : Posterior (Bayes)
# -------------------------------
P_M_given_R = (P_R_M * P_M) / P_R
P_B_given_R = (P_R_B * P_B) / P_R


# =============================================================================
# AFFICHAGE DES RÉSULTATS
# =============================================================================

print("===============================================")
print("THÉORÈME DE BAYES - INTERPRÉTATION")
print("===============================================")

print("\n1- PRIORS")
print(f"P(Maligne) = {P_M:.3f}")
print(f"P(Bénigne) = {P_B:.3f}")

print("\n2- VRAISEMBLANCES")
print(f"P(radius > {seuil} | Maligne) = {P_R_M:.3f}")
print(f"P(radius > {seuil} | Bénigne) = {P_R_B:.3f}")

print("\n3- ÉVIDENCE")
print(f"P(radius > {seuil}) = {P_R:.3f}")

print("\n4- POSTERIOR")
print(f"P(Maligne | radius > {seuil}) = {P_M_given_R:.3f}")
print(f"P(Bénigne | radius > {seuil}) = {P_B_given_R:.3f}")

print("\nINTERPRÉTATION")
print(f"Si radius_mean > {seuil}, la probabilité de cancer devient {P_M_given_R*100:.1f}%")
```

```python
# ============================================================
# NAIVE BAYES - Classification du cancer du sein
# Implémentation sklearn + version from scratch
# ============================================================

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np

# ------------------------------------------------------------
# 1. Modèle Naive Bayes avec sklearn
# ------------------------------------------------------------

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

y_pred_nb = nb_model.predict(X_test)

accuracy_nb = accuracy_score(y_test, y_pred_nb)
cm_nb = confusion_matrix(y_test, y_pred_nb)

print("NAIVE BAYES - SKLEARN")
print("Accuracy :", accuracy_nb)

print("\nConfusion Matrix")
print(cm_nb)

print("\nClassification Report")
print(classification_report(y_test, y_pred_nb))


# ------------------------------------------------------------
# 2. Naive Bayes From Scratch
# ------------------------------------------------------------

class GaussianNaiveBayes:

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.priors = {}
        self.means = {}
        self.vars = {}

        for c in self.classes:
            Xc = X[y == c]
            self.priors[c] = len(Xc) / len(X)
            self.means[c] = Xc.mean(axis=0)
            self.vars[c] = Xc.var(axis=0) + 1e-9

    def gaussian_log(self, X, mean, var):
        return -0.5 * (np.log(2*np.pi*var) + (X-mean)**2 / var)

    def predict(self, X):
        log_probs = []

        for c in self.classes:
            log_prior = np.log(self.priors[c])
            log_likelihood = self.gaussian_log(X, self.means[c], self.vars[c]).sum(axis=1)
            log_probs.append(log_prior + log_likelihood)

        log_probs = np.array(log_probs).T
        return self.classes[np.argmax(log_probs, axis=1)]


# entraînement
nb_scratch = GaussianNaiveBayes()
nb_scratch.fit(X_train, y_train)

y_pred_scratch = nb_scratch.predict(X_test)

accuracy_scratch = accuracy_score(y_test, y_pred_scratch)

print("\nNAIVE BAYES - FROM SCRATCH")
print("Accuracy :", accuracy_scratch)
```

## **6- KNN - Optimisation et Entraînement**

```python
# ══════════════════════════════════════════════════════════════
#  KNN — IMPLÉMENTATION FROM SCRATCH
# ══════════════════════════════════════════════════════════════

from collections import Counter
from sklearn.model_selection import StratifiedKFold, cross_val_score

class KNNClassifier:
    """
    K-Nearest Neighbors from scratch.
    Distance euclidienne vectorisée (pas de boucle sur le train).
    """
    def __init__(self, k=5):
        self.k = k

    def fit(self, X, y):
        self.X_tr = X
        self.y_tr = y
        return self

    def _distances(self, x):
        """Distance euclidienne entre x et tous les points du train."""
        return np.sqrt(np.sum((self.X_tr - x)**2, axis=1))

    def predict_single(self, x):
        dists  = self._distances(x)
        k_idx  = np.argsort(dists)[:self.k]
        k_labs = self.y_tr[k_idx]
        return Counter(k_labs).most_common(1)[0][0]

    def predict(self, X):
        return np.array([self.predict_single(x) for x in X])


# ── Cross-validation pour trouver k optimal ───────────────────
k_range     = range(1, 22, 2)   # k impairs uniquement (évite les égalités)
cv_means    = []
cv_stds     = []
cv_splitter = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for k in k_range:
    knn_cv = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn_cv, X_train_s, y_train, cv=cv_splitter, scoring='accuracy')
    cv_means.append(scores.mean())
    cv_stds.append(scores.std())

k_optimal = list(k_range)[np.argmax(cv_means)]

print('═'*55)
print('  RECHERCHE DU K OPTIMAL (CV 5-fold, k impairs)')
print('═'*55)
print(f'  {"k":<6} {"Acc. moy":>10} {"±Std":>8}')
print('  ' + '-'*26)
for k, m, s in zip(k_range, cv_means, cv_stds):
    star = '  ← OPTIMAL' if k == k_optimal else ''
    print(f'  k={k:<4} {m*100:>9.2f}%  ±{s*100:.2f}%{star}')
print(f'\n  → k retenu : k = {k_optimal}')
```

```python
# ── Entraînement final KNN ─────────────────────────────────────
knn_final  = KNeighborsClassifier(n_neighbors=k_optimal).fit(X_train_s, y_train)
knn_sc     = KNNClassifier(k=k_optimal).fit(X_train_s, y_train)

y_pred_knn_sc = knn_sc.predict(X_test_s)
y_pred_knn_sk = knn_final.predict(X_test_s)
y_proba_knn   = knn_final.predict_proba(X_test_s)[:, 1]

acc_knn_sc = accuracy_score(y_test, y_pred_knn_sc)
acc_knn_sk = accuracy_score(y_test, y_pred_knn_sk)

print('═'*55)
print(f'  KNN (k={k_optimal}) — RÉSULTATS FINAUX')
print('═'*55)
print(f'  Accuracy (from scratch) : {acc_knn_sc*100:.2f}%')
print(f'  Accuracy (sklearn)      : {acc_knn_sk*100:.2f}%')
print(f'  → Convergence parfaite')
print()
print(classification_report(y_test, y_pred_knn_sk, target_names=['Maligne(0)', 'Bénigne(1)']))
```

```python
# ══════════════════════════════════════════════════════════════
#  MÉTRIQUES COMPLÈTES
# ══════════════════════════════════════════════════════════════

def metriques(y_true, y_pred, y_proba, nom):
    cm   = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    acc  = accuracy_score(y_true, y_pred)
    sens = tp / (tp + fn)              # Sensibilité (rappel Bénigne)
    spec = tn / (tn + fp)              # Spécificité (rappel Maligne)
    prec = tp / (tp + fp) if (tp+fp)>0 else 0
    f1   = 2*prec*sens / (prec+sens) if (prec+sens)>0 else 0
    auc  = roc_auc_score(y_true, y_proba)
    fn_count = fn
    return {'Modèle': nom, 'Accuracy': acc, 'Sensibilité': sens,
            'Spécificité': spec, 'Précision': prec,
            'F1-Score': f1, 'AUC-ROC': auc, 'Faux Négatifs': fn_count}

m_nb  = metriques(y_test, y_pred_nb,  y_proba_nb,  'Naive Bayes')
m_knn = metriques(y_test, y_pred_knn_sk, y_proba_knn, f'KNN (k={k_optimal})')

df_comp = pd.DataFrame([m_nb, m_knn]).set_index('Modèle')

print('═'*75)
print('  TABLEAU COMPARATIF FINAL')
print('═'*75)
cols_pct = ['Accuracy','Sensibilité','Spécificité','Précision','F1-Score']
cols_raw = ['AUC-ROC']
header = f'  {"Modèle":<20}'
for c in cols_pct + cols_raw:
    header += f'{c:>13}'
print(header)
print('  ' + '-'*73)
for model in df_comp.index:
    row = f'  {model:<20}'
    for c in cols_pct:
        v = df_comp.loc[model, c]
        best = df_comp[c].idxmax()
        mark = '' if model == best else ' '
        row += f'{v*100:>10.2f}%{mark:>2}'
    for c in cols_raw:
        v = df_comp.loc[model, c]
        best = df_comp[c].idxmax()
        mark = '' if model == best else ' '
        row += f'{v:>12.4f}{mark:>1}'
    print(row)
print('═'*75)
print(f'\n  Faux Négatifs (maligne classée bénigne — DANGER CLINIQUE) :')
for model in df_comp.index:
    fn = df_comp.loc[model, 'Faux Négatifs']
    print(f'    {model:<22} : {fn} faux négatifs ')
```

# **7. Arbre de Décision (Decision Tree)**

## Principe

Un **Arbre de Décision** construit un ensemble de règles de type `SI... ALORS...` à partir des données.  
À chaque nœud, l'arbre pose une question sur une feature (ex : *"radius_mean > 15 ?"*) et choisit le **seuil qui sépare le mieux les deux classes**, mesuré par l'**impureté de Gini** :

$$Gini = 1 - \sum_{k} p_k^2$$

Plus Gini est faible, plus le nœud est *pur* (une seule classe domine).  
L'arbre s'arrête quand les feuilles sont pures ou quand on atteint la profondeur maximale fixée.

## Avantages
- Complètement interprétable : on peut lire les règles de décision
- Pas besoin de normalisation des données
- Gère naturellement les corrélations entre features

## Inconvénient
- Tendency au surapprentissage (overfitting) si la profondeur n'est pas contrôlée


```python
# ══════════════════════════════════════════════════════════════
#  DECISION TREE — Entraînement et optimisation
# ══════════════════════════════════════════════════════════════

from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier # Import DecisionTreeClassifier

# ── Recherche de la profondeur optimale ──────────────────────
# On teste différentes profondeurs max et on garde celle qui
# donne le meilleur score en cross-validation 5-fold

depths      = range(1, 16)
cv_scores   = []

for d in depths:
    dt = DecisionTreeClassifier(max_depth=d, random_state=42, criterion='gini')
    scores = cross_val_score(dt, X_train_s, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())

best_depth = depths[cv_scores.index(max(cv_scores))]

print('═'*60)
print('  ARBRE DE DÉCISION — Recherche de la profondeur optimale')
print('═'*60)
for d, s in zip(depths, cv_scores):
    marker = '  ← OPTIMAL' if d == best_depth else ''
    print(f'  depth={d:2d}  →  CV Accuracy = {s*100:.2f}%{marker}')
print(f'\n  Profondeur optimale retenue : {best_depth}')

# ── Tracé de la courbe ───────────────────────────────────────
plt.figure(figsize=(10, 4))
plt.plot(list(depths), [s*100 for s in cv_scores],
         color='#2563EB', lw=2.5, marker='o', markersize=7)
plt.axvline(best_depth, color='#16A34A', lw=2.5, ls='--',
            label=f'Profondeur optimale = {best_depth}')
plt.scatter([best_depth], [max(cv_scores)*100], color='#16A34A', s=150, zorder=5)
plt.xlabel('Profondeur maximale (max_depth)', fontsize=12)
plt.ylabel('Accuracy (CV 5-fold, %)', fontsize=12)
plt.title('Arbre de Décision — Recherche de la profondeur optimale', fontsize=13, fontweight='bold')
plt.legend(fontsize=11); plt.grid(alpha=0.3); plt.tight_layout()
plt.show()
print(f"Interprétation : pour depth < {best_depth}, l'arbre est trop peu profond (underfitting).")
print(f'Pour depth > {best_depth}, il commence à mémoriser le bruit (overfitting).')
```

# **8. Forêt Aléatoire (Random Forest)**

## Principe

Une **Forêt Aléatoire** est un **ensemble** de nombreux arbres de décision entraînés chacun sur :
- Un **sous-échantillon aléatoire** des données (bootstrap)
- Un **sous-ensemble aléatoire** des features à chaque nœud

La prédiction finale = **vote majoritaire** de tous les arbres.

$$\hat{y} = \text{mode}\left(\hat{y}_1, \hat{y}_2, ..., \hat{y}_n\right)$$

## Pourquoi c'est mieux qu'un seul arbre ?
| | Arbre seul | Forêt Aléatoire |
|---|---|---|
| Overfitting | Élevé | Faible (arbres se compensent) |
| Stabilité | Faible | Élevée |
| Interprétabilité | Facile | Plus difficile |
| Performance | Moyenne | Généralement meilleure |

Le principe clé est la **décorrélation** : comme chaque arbre voit des données et des features différentes, leurs erreurs ne sont pas corrélées. En moyennant, les erreurs s'annulent.


```python
# ══════════════════════════════════════════════════════════════
#  RANDOM FOREST — Entraînement et optimisation
# ══════════════════════════════════════════════════════════════

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# ── Recherche du nombre d'arbres optimal ─────────────────────
# On teste différentes valeurs de n_estimators (nombre d'arbres)
# et on observe la stabilisation de l'accuracy.

n_trees_range = [10, 25, 50, 75, 100, 150, 200, 300, 500]
rf_cv_scores  = []

for n in n_trees_range:
    rf = RandomForestClassifier(n_estimators=n, random_state=42, n_jobs=-1)
    scores = cross_val_score(rf, X_train_s, y_train, cv=5, scoring='accuracy')
    rf_cv_scores.append(scores.mean())

best_n = n_trees_range[rf_cv_scores.index(max(rf_cv_scores))]

print('═'*60)
print('  RANDOM FOREST — Recherche du nombre d\'arbres optimal')
print('═'*60)
for n, s in zip(n_trees_range, rf_cv_scores):
    marker = '  ← OPTIMAL' if n == best_n else ''
    print(f'  n_estimators={n:4d}  →  CV Accuracy = {s*100:.2f}%{marker}')
print(f'\n  Nombre d\'arbres optimal retenu : {best_n}')

# ── Courbe de convergence ─────────────────────────────────────
plt.figure(figsize=(10, 4))
plt.plot(n_trees_range, [s*100 for s in rf_cv_scores],
         color='#16A34A', lw=2.5, marker='o', markersize=8)
plt.axvline(best_n, color='#C0392B', lw=2.5, ls='--',
            label=f'n optimal = {best_n} arbres')
plt.scatter([best_n], [max(rf_cv_scores)*100], color='#C0392B', s=150, zorder=5)
plt.xlabel('Nombre d\'arbres (n_estimators)', fontsize=12)
plt.ylabel('Accuracy (CV 5-fold, %)', fontsize=12)
plt.title('Random Forest — Convergence avec le nombre d\'arbres\n'
          '(L\'accuracy se stabilise : ajouter plus d\'arbres n\'aide plus)', fontsize=12, fontweight='bold')
plt.legend(fontsize=11); plt.grid(alpha=0.3); plt.tight_layout()
plt.show()
print('Observation : l\'accuracy se stabilise rapidement.')
print('Au-delà d\'un certain nombre d\'arbres, on n\'améliore plus.')
print('C\'est la loi des rendements décroissants de l\'ensemble.')
```

# 10. K-Means — Clustering et selection automatique de k optimal

## Principe

Le K-Means partitionne les donnees en k groupes en minimisant l'inertie
intra-cluster (somme des distances au centre du cluster) :

$$\text{Inertie} = \sum_{i=1}^{k} \sum_{x \in C_i} \| x - \mu_i \|^2$$

Nous l'appliquons sur les donnees normalisees **sans utiliser les etiquettes**
(apprentissage non supervise) pour verifier si les donnees contiennent une
structure naturelle en 2 groupes correspondant aux deux classes cliniques.

Pour choisir k, nous utilisons deux criteres complementaires :
- **Methode du coude (Elbow)** : on cherche le k ou la baisse d'inertie ralentit
- **Silhouette Score** : on cherche le k qui maximise la coherence des clusters


```python
# ============================================================
# ETAPE 10 — K-MEANS : SELECTION AUTOMATIQUE DE K OPTIMAL
# ============================================================
#
# Principe de la selection de k :
#   - On entraine K-Means pour k allant de 2 a 10
#   - On calcule l'inertie et le Silhouette Score pour chaque k
#   - Le coude de la courbe d'inertie indique le k naturel
#   - Le pic du Silhouette Score confirme le choix
#
# Note : on travaille sur X_test_s (normalise) pour
# evaluer la separabilite sur des donnees non vues.
# ============================================================

k_range    = range(2, 11)
inerties   = []
silhouettes = []

print('='*60)
print('  ETAPE 10 — K-MEANS : RECHERCHE DU K OPTIMAL')
print('='*60)
print()
print(f'  {"k":>4}  {"Inertie":>12}  {"Silhouette":>12}')
print('  ' + '-'*32)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_km = km.fit_predict(X_test_s)
    inertie   = km.inertia_
    sil       = silhouette_score(X_test_s, labels_km)
    inerties.append(inertie)
    silhouettes.append(sil)
    print(f'  {k:>4}  {inertie:>12.2f}  {sil:>12.4f}')

k_optimal_km = list(k_range)[silhouettes.index(max(silhouettes))]
print()
print(f'  k optimal (Silhouette max) : {k_optimal_km}')

# ── Graphiques de selection ───────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('K-Means — Selection du nombre de clusters optimal',
             fontsize=13, fontweight='bold')

# Coude
axes[0].plot(list(k_range), inerties, '#2563EB', lw=2.5, marker='o', markersize=8)
axes[0].axvline(k_optimal_km, color='#C0392B', lw=2, ls='--',
                label=f'k optimal = {k_optimal_km}')
axes[0].scatter([k_optimal_km], [inerties[k_optimal_km-2]],
                color='#C0392B', s=150, zorder=5)
axes[0].set_xlabel('Nombre de clusters k', fontsize=12)
axes[0].set_ylabel('Inertie intra-cluster', fontsize=12)
axes[0].set_title('Methode du Coude (Elbow Method)\n'
                  'Le coude indique le k naturel dans les donnees', fontweight='bold')
axes[0].legend(fontsize=11); axes[0].grid(alpha=0.25)

# Silhouette
axes[1].plot(list(k_range), silhouettes, '#16A34A', lw=2.5, marker='s', markersize=8)
axes[1].axvline(k_optimal_km, color='#C0392B', lw=2, ls='--',
                label=f'k optimal = {k_optimal_km}')
axes[1].scatter([k_optimal_km], [max(silhouettes)], color='#C0392B', s=150, zorder=5)
axes[1].set_xlabel('Nombre de clusters k', fontsize=12)
axes[1].set_ylabel('Silhouette Score', fontsize=12)
axes[1].set_title('Silhouette Score par k\n'
                  'Plus il est proche de 1, meilleure est la cohesion', fontweight='bold')
axes[1].legend(fontsize=11); axes[1].grid(alpha=0.25)

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# ETAPE 10.2 — K-MEANS : EVALUATION ET VISUALISATION
# ============================================================
#
# On entraine le K-Means final avec k optimal.
# On visualise les clusters en 2D (PCA) et on mesure
# leur alignement avec les vraies etiquettes cliniques.
#
# Attention : K-Means est non supervise. Les numeros
# de clusters n'ont pas de correspondance directe
# avec Maligne/Benigne — on aligne manuellement.
# ============================================================

km_final = KMeans(n_clusters=k_optimal_km, random_state=42, n_init=10)
labels_km_final = km_final.fit_predict(X_test_s)

# Alignement des labels de cluster avec les vraies classes
# (le cluster majoritaire prend l'etiquette de la classe dominante)
from scipy.stats import mode as scipy_mode
mapping = {}
for cluster_id in range(k_optimal_km):
    mask      = labels_km_final == cluster_id
    true_labs = y_test[mask]
    if len(true_labs) > 0:
        dom_class = int(scipy_mode(true_labs, keepdims=True).mode[0])
        mapping[cluster_id] = dom_class

labels_km_aligned = np.array([mapping.get(l, 0) for l in labels_km_final])

# Metriques de clustering
sil_final = silhouette_score(X_test_s, labels_km_final)
db_final  = davies_bouldin_score(X_test_s, labels_km_final)
acc_km    = accuracy_score(y_test, labels_km_aligned)

print('='*60)
print('  K-MEANS FINAL — Resultats')
print('='*60)
print(f'  k utilise            : {k_optimal_km}')
print(f'  Silhouette Score     : {sil_final:.4f}  (1=parfait, 0=clusters chevauchants)')
print(f'  Davies-Bouldin Index : {db_final:.4f}  (0=parfait, plus faible=meilleur)')
print(f'  Accuracy (aligne)    : {acc_km*100:.2f}%  (correspondance cluster/classe reelle)')
print()
print(f'  Interpretation :')
print(f'  Un Silhouette de {sil_final:.3f} indique que les clusters sont')
if sil_final > 0.5:
    print('  bien separes et coherents — la structure en 2 classes')
    print('  est clairement visible dans les donnees.')
elif sil_final > 0.25:
    print('  moderement separes — une structure existe mais avec')
    print('  chevauchement partiel entre les classes.')
else:
    print('  faiblement separes — les classes se chevauchent')
    print('  dans l espace des 30 features.')

# ── Visualisation PCA 2D ──────────────────────────────────────
pca2 = PCA(n_components=2, random_state=42)
X_2d = pca2.fit_transform(X_test_s)

fig, axes = plt.subplots(1, 3, figsize=(17, 5))
fig.suptitle('K-Means — Visualisation 2D (PCA) des clusters',
             fontsize=13, fontweight='bold')

# Clusters K-Means
colors_km = ['#2563EB', '#C0392B', '#16A34A', '#F59E0B']
for cid in range(k_optimal_km):
    mask = labels_km_final == cid
    axes[0].scatter(X_2d[mask, 0], X_2d[mask, 1],
                    color=colors_km[cid % len(colors_km)],
                    alpha=0.65, s=35, label=f'Cluster {cid}', edgecolors='none')

# Centres des clusters projetes
centers_2d = pca2.transform(km_final.cluster_centers_)
axes[0].scatter(centers_2d[:, 0], centers_2d[:, 1],
                color='black', s=200, marker='X', zorder=5, label='Centres')
axes[0].set_title('Clusters K-Means', fontweight='bold')
axes[0].set_xlabel(f'PC1 ({pca2.explained_variance_ratio_[0]*100:.1f}%)')
axes[0].set_ylabel(f'PC2 ({pca2.explained_variance_ratio_[1]*100:.1f}%)')
axes[0].legend(fontsize=9); axes[0].grid(alpha=0.2)

# Vraies etiquettes
for cls, col, lbl in [(0,'#EF4444','Maligne'),(1,'#22C55E','Benigne')]:
    mask = y_test == cls
    axes[1].scatter(X_2d[mask, 0], X_2d[mask, 1],
                    color=col, alpha=0.65, s=35, label=lbl, edgecolors='none')
axes[1].set_title('Vraies etiquettes cliniques', fontweight='bold')
axes[1].set_xlabel(f'PC1 ({pca2.explained_variance_ratio_[0]*100:.1f}%)')
axes[1].set_ylabel(f'PC2 ({pca2.explained_variance_ratio_[1]*100:.1f}%)')
axes[1].legend(fontsize=10); axes[1].grid(alpha=0.2)

# Silhouette plot par echantillon
sil_samples = silhouette_samples(X_test_s, labels_km_final)
y_lower = 10
for cid in range(k_optimal_km):
    sil_vals = np.sort(sil_samples[labels_km_final == cid])
    size_c   = sil_vals.shape[0]
    y_upper  = y_lower + size_c
    axes[2].fill_betweenx(np.arange(y_lower, y_upper), 0, sil_vals,
                           color=colors_km[cid % len(colors_km)], alpha=0.75)
    axes[2].text(-0.05, y_lower + size_c/2, f'C{cid}', fontsize=10)
    y_lower  = y_upper + 10

axes[2].axvline(sil_final, color='#C0392B', lw=2, ls='--',
                label=f'Moyenne = {sil_final:.3f}')
axes[2].set_title('Silhouette par echantillon\n'
                  'Chaque barre = 1 patient, largeur = coherence',
                  fontweight='bold')
axes[2].set_xlabel('Coefficient de silhouette')
axes[2].set_ylabel('Patients (groupes par cluster)')
axes[2].legend(fontsize=9); axes[2].grid(axis='x', alpha=0.25)

plt.tight_layout()
plt.show()
```

# 11. Clustering Hierarchique — Dendrogramme et validation

## Principe

Le clustering hierarchique construit une arborescence (dendrogramme) en
fusionnant iterativement les paires de clusters les plus proches.
Contrairement a K-Means, il ne necessite pas de specifier k a priori.

La distance entre clusters est mesuree par un **critere de liaison** :
- **Ward** : minimise la variance intra-cluster apres fusion (recommande)
- **Complete** : distance maximale entre points de deux clusters
- **Average** : distance moyenne entre tous les paires de points

On coupe le dendrogramme a la hauteur qui maximise le Silhouette Score.


```python
# ============================================================
# ETAPE 11 — CLUSTERING HIERARCHIQUE + DENDROGRAMME
# ============================================================
#
# Methode : linkage de Ward sur X_test_s
# Ward minimise la somme des carres des distances intra-cluster
# apres chaque fusion — c'est le critere le mieux adapte
# aux donnees de taille comparable comme notre dataset.
#
# Le dendrogramme montre la hierarchie des fusions.
# La hauteur de chaque noeud = distance de la fusion.
# On identifie le saut le plus grand = k naturel.
# ============================================================

# Construction de la matrice de liaison
Z = linkage(X_test_s, method='ward')

fig, axes = plt.subplots(1, 2, figsize=(17, 7))
fig.suptitle('Clustering Hierarchique (methode de Ward) — Dataset Cancer',
             fontsize=13, fontweight='bold')

# ── Dendrogramme complet ──────────────────────────────────────
dendrogram(
    Z,
    ax             = axes[0],
    truncate_mode  = 'lastp',   # n'afficher que les 20 dernieres fusions
    p              = 20,
    leaf_rotation  = 90,
    leaf_font_size = 9,
    show_contracted= True,
    color_threshold= 0.7 * max(Z[:, 2])
)
axes[0].set_title('Dendrogramme (20 dernieres fusions affichees)\n'
                  'La hauteur d une barre = distance de la fusion',
                  fontweight='bold')
axes[0].set_xlabel('Patients (ou sous-groupes fusionnes)')
axes[0].set_ylabel('Distance de Ward')
axes[0].axhline(0.7 * max(Z[:, 2]), color='#C0392B', lw=2, ls='--',
                label='Seuil de coupure')
axes[0].legend(fontsize=10)
axes[0].grid(axis='y', alpha=0.2)

# ── Selection du nombre de clusters par Silhouette ────────────
sil_hiera = []
k_range_h = range(2, 8)

for k in k_range_h:
    labels_h = fcluster(Z, k, criterion='maxclust')
    sil_h    = silhouette_score(X_test_s, labels_h)
    sil_hiera.append(sil_h)

k_opt_h = list(k_range_h)[sil_hiera.index(max(sil_hiera))]

axes[1].plot(list(k_range_h), sil_hiera, '#7C3AED', lw=2.5, marker='o', markersize=9)
axes[1].axvline(k_opt_h, color='#C0392B', lw=2, ls='--',
                label=f'k optimal = {k_opt_h}')
axes[1].scatter([k_opt_h], [max(sil_hiera)], color='#C0392B', s=150, zorder=5)
axes[1].set_xlabel('Nombre de clusters k', fontsize=12)
axes[1].set_ylabel('Silhouette Score', fontsize=12)
axes[1].set_title('Silhouette Score vs nombre de clusters\n(Clustering Hierarchique Ward)',
                  fontweight='bold')
axes[1].legend(fontsize=11); axes[1].grid(alpha=0.25)

plt.tight_layout()
plt.show()

# ── Evaluation finale ─────────────────────────────────────────
labels_hiera = fcluster(Z, k_opt_h, criterion='maxclust') - 1  # 0-indexed
sil_h_final  = silhouette_score(X_test_s, labels_hiera)
db_h_final   = davies_bouldin_score(X_test_s, labels_hiera)

print('='*60)
print('  CLUSTERING HIERARCHIQUE — Resultats')
print('='*60)
print(f'  Methode de liaison    : Ward')
print(f'  k optimal retenu      : {k_opt_h}')
print(f'  Silhouette Score      : {sil_h_final:.4f}')
print(f'  Davies-Bouldin Index  : {db_h_final:.4f}')
print()
print(f'  Comparaison K-Means vs Hierarchique :')
print(f'  {"Metrique":<25}  {"K-Means":>10}  {"Hierarchique":>14}')
print('  ' + '-'*52)
print(f'  {"Silhouette Score":<25}  {sil_final:>10.4f}  {sil_h_final:>14.4f}')
print(f'  {"Davies-Bouldin Index":<25}  {db_final:>10.4f}  {db_h_final:>14.4f}')
print()
if sil_h_final > sil_final:
    print('  Le clustering hierarchique produit des clusters plus coherents.')
else:
    print('  Le K-Means produit des clusters plus coherents sur ce dataset.')
print()
print('  Interpretation du Davies-Bouldin Index :')
print('  Proche de 0 = clusters compacts et bien separes.')
print('  Valeur elevee = clusters disperses ou qui se chevauchent.')
```

# 12. Metriques de Similarite Structurelle — Dice, ASD, SSIM

## Dice Score

Le **Dice Score** (ou coefficient de Sorensen-Dice) mesure le chevauchement
entre deux ensembles. En classification binaire, il est equivalent au F1-Score :

$$\text{Dice} = \frac{2 \cdot |A \cap B|}{|A| + |B|} = \frac{2 \cdot TP}{2 \cdot TP + FP + FN}$$

## Average Symmetric Surface Distance (ASD)

L'**ASD** mesure la distance moyenne entre les surfaces de deux ensembles.
Adapte ici aux predictions : on mesure la distance moyenne entre les
probabilites predites et les vraies classes dans l'espace des scores.

$$\text{ASD}(A, B) = \frac{d(A, B) + d(B, A)}{2}$$

## SSIM — Structural Similarity Index Measure

Le **SSIM** evalue la similarite structurelle entre deux signaux.
Initialement concu pour les images, on l'adapte ici aux vecteurs de
probabilites predites : on compare la distribution des scores de chaque
modele a la distribution ideale des vraies etiquettes.


```python
# ============================================================
# ETAPE 12 — DICE SCORE, ASD, SSIM
# ============================================================
#
# Ces trois metriques evaluent la similarite structurelle
# entre les predictions d'un modele et la realite.
#
# Dice  : chevauchement ensembliste (identique a F1)
# ASD   : distance moyenne entre frontieres de decision
# SSIM  : similarite de la distribution des scores predits
#         vs distribution ideale (vraies etiquettes)
# ============================================================

def dice_score(y_true, y_pred):
    """
    Coefficient de Sorensen-Dice pour classification binaire.
    Identique au F1-Score : Dice = 2TP / (2TP + FP + FN).
    Varie entre 0 (aucun chevauchement) et 1 (correspondance parfaite).
    """
    tp = int(((y_pred == 1) & (y_true == 1)).sum())
    fp = int(((y_pred == 1) & (y_true == 0)).sum())
    fn = int(((y_pred == 0) & (y_true == 1)).sum())
    denom = 2 * tp + fp + fn
    return (2 * tp) / denom if denom > 0 else 0.0


def asd_score(y_true_proba, y_pred_proba):
    """
    Average Symmetric Surface Distance adapte aux probabilites.

    On interprete les vecteurs de probabilites comme deux 'surfaces'
    dans l'espace 1D des scores. L'ASD mesure la distance moyenne
    entre les distributions de probabilites predites et ideales.

    Methode :
      - On trie les deux vecteurs
      - On calcule la distance point-a-point entre les distributions
        empiriques cumulees (CDF)
      - L'ASD = distance moyenne entre les deux CDF
    """
    a = np.sort(y_true_proba.astype(float))
    b = np.sort(y_pred_proba.astype(float))
    # Interpolation pour aligner les deux vecteurs
    n   = max(len(a), len(b))
    a_i = np.interp(np.linspace(0, 1, n), np.linspace(0, 1, len(a)), a)
    b_i = np.interp(np.linspace(0, 1, n), np.linspace(0, 1, len(b)), b)
    return float(np.mean(np.abs(a_i - b_i)))


def ssim_1d(signal_a, signal_b, L=1.0):
    """
    SSIM adapte aux vecteurs 1D de probabilites.

    Formule originale SSIM (Wang et al. 2004) :
      SSIM(x,y) = (2*mu_x*mu_y + C1)(2*sigma_xy + C2)
                  / (mu_x^2 + mu_y^2 + C1)(sigma_x^2 + sigma_y^2 + C2)

    Ici on l'applique sur les distributions de scores predits.
    L = plage dynamique des valeurs (1.0 pour des probabilites).
    C1 = (k1*L)^2, C2 = (k2*L)^2 avec k1=0.01, k2=0.03 (standard).
    """
    k1, k2   = 0.01, 0.03
    C1, C2   = (k1 * L)**2, (k2 * L)**2

    mu_a     = np.mean(signal_a)
    mu_b     = np.mean(signal_b)
    sigma_a2 = np.var(signal_a)
    sigma_b2 = np.var(signal_b)
    sigma_ab = np.mean((signal_a - mu_a) * (signal_b - mu_b))

    numerateur   = (2 * mu_a * mu_b + C1) * (2 * sigma_ab + C2)
    denominateur = (mu_a**2 + mu_b**2 + C1) * (sigma_a2 + sigma_b2 + C2)
    return float(numerateur / denominateur) if denominateur != 0 else 0.0


# Distribution ideale (les vraies etiquettes comme probabilites parfaites)
y_ideal = y_test.astype(float)

# ── Calcul pour tous les modeles ─────────────────────────────
modeles_sim = {
    'Naive Bayes'           : (y_pred_nb,     y_proba_nb),
    f'KNN (k={k_optimal})'  : (y_pred_knn_sk, y_proba_knn),
    f'DT (depth={best_depth})': (y_pred_dt,   y_proba_dt),
    f'RF ({best_n} arbres)' : (y_pred_rf,     y_proba_rf),
}

print('='*70)
print('  ETAPE 12 — DICE SCORE, ASD, SSIM — TOUS MODELES')
print('='*70)
print()
print(f'  {"Modele":<25}  {"Dice":>8}  {"ASD":>10}  {"SSIM":>8}')
print('  ' + '-'*55)

dice_vals = {}; asd_vals = {}; ssim_vals = {}

for nom, (y_pred_m, y_proba_m) in modeles_sim.items():
    d  = dice_score(y_test, y_pred_m)
    a  = asd_score(y_ideal, y_proba_m)
    s  = ssim_1d(y_ideal, y_proba_m)
    dice_vals[nom] = d
    asd_vals[nom]  = a
    ssim_vals[nom] = s
    print(f'  {nom:<25}  {d:>8.4f}  {a:>10.6f}  {s:>8.4f}')

print()
print('  Dice  : identique au F1-Score (chevauchement ensembliste).')
print('  ASD   : distance moyenne entre la distribution predite et ideale.')
print('          Plus il est proche de 0, plus les scores sont calibres.')
print('  SSIM  : similarite structurelle entre la distribution des scores')
print('          predits et les vraies etiquettes. Proche de 1 = structure')
print('          identique (memes moyennes, variances et correlations).')

# ── Visualisation ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(17, 5))
fig.suptitle('Metriques de Similarite Structurelle — Dice, ASD, SSIM',
             fontsize=13, fontweight='bold')

colors_m = ['#2563EB','#16A34A','#F59E0B','#C0392B']
noms     = list(modeles_sim.keys())
noms_court = ['NB','KNN','DT','RF']

for ax, (vals, titre, desc, inverse) in zip(axes[:3], [
    (dice_vals, 'Dice Score',  'Plus proche de 1 = meilleur', False),
    (asd_vals,  'ASD',         'Plus proche de 0 = meilleur', True),
    (ssim_vals, 'SSIM',        'Plus proche de 1 = meilleur', False),
]):
    v = [vals[n] for n in noms]
    bars = ax.bar(noms_court, v, color=colors_m, alpha=0.85, edgecolor='white', width=0.55)
    ax.set_title(f'{titre}\n{desc}', fontweight='bold')
    ax.grid(axis='y', alpha=0.25)
    for bar, val in zip(bars, v):
        ax.text(bar.get_x() + bar.get_width()/2, val + max(v)*0.01,
                f'{val:.4f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

# 13. DCG — Discounted Cumulative Gain

## Principe

Le **DCG** (Discounted Cumulative Gain) est une metrique de classement
(ranking). L'idee est simple : un bon modele place les vrais positifs
(tumeurs benignes) **en haut de son classement** par score de confiance.

$$\text{DCG}@k = \sum_{i=1}^{k} \frac{\text{rel}_i}{\log_2(i+1)}$$

- $\text{rel}_i$ = pertinence du i-eme element (1 si benigne, 0 si maligne)
- Le denominateur $\log_2(i+1)$ penalise les positifs trouves tard

Le **nDCG** (normalise) divise par le DCG ideal (tous les positifs en premier),
ce qui donne une valeur entre 0 et 1 comparable entre datasets.


```python
# ============================================================
# ETAPE 13 — DCG ET nDCG (DISCOUNTED CUMULATIVE GAIN)
# ============================================================
#
# Algorithme :
#   1. Trier les patients par score de confiance decroissant
#   2. Calculer le gain de trouver un vrai positif a la position i
#   3. Ponderer par 1/log2(i+1) — les positions tardives valent moins
#   4. Normaliser par le gain ideal (tous les positifs d'abord)
#
# Interpretation clinique :
#   Un nDCG@10 = 0.95 signifie que parmi les 10 patients
#   que le modele classe comme les "plus probablement benignes",
#   95% de l'information pertinente est presente.
# ============================================================

def dcg_at_k(y_true, y_scores, k=None):
    """
    Discounted Cumulative Gain a la position k.
    On trie par score decroissant et on somme rel_i / log2(i+1).
    """
    order    = np.argsort(y_scores)[::-1]
    y_sorted = np.array(y_true)[order]
    if k is not None:
        y_sorted = y_sorted[:k]
    positions = np.arange(2, len(y_sorted) + 2)   # positions 1,2,3,...
    gains     = y_sorted / np.log2(positions)
    return float(gains.sum())


def ndcg_at_k(y_true, y_scores, k=None):
    """
    nDCG normalise. Le DCG ideal assume tous les positifs en tete.
    nDCG = DCG_obtenu / DCG_ideal.
    """
    dcg_val   = dcg_at_k(y_true, y_scores, k)
    ideal_scores = np.ones_like(y_scores)   # modele parfait : tous positifs d'abord
    ideal_val = dcg_at_k(y_true, ideal_scores, k)
    return dcg_val / ideal_val if ideal_val > 0 else 0.0


k_values_dcg = [5, 10, 20, 50, len(y_test)]

print('='*75)
print('  ETAPE 13 — nDCG@k — DISCOUNTED CUMULATIVE GAIN NORMALISE')
print('='*75)
print()
header = f'  {"Modele":<25}'
for k in k_values_dcg:
    lbl = f'@{k}' if k < len(y_test) else '@all'
    header += f'  {lbl:>9}'
print(header)
print('  ' + '-'*70)

ndcg_results = {}

for nom, (_, y_proba_m) in modeles_sim.items():
    row = f'  {nom:<25}'
    ndcg_results[nom] = []
    for k in k_values_dcg:
        kk  = None if k == len(y_test) else k
        val = ndcg_at_k(y_test, y_proba_m, k=kk)
        ndcg_results[nom].append(val)
        row += f'  {val:>9.4f}'
    print(row)

print()
print('  nDCG@k = 1.0 signifie que les k premiers patients classes comme')
print('  les plus probablement benignes sont tous effectivement benignes.')
print('  Proche de 1 au petit k = le modele est tres confiant sur ses')
print('  meilleures predictions.')

# ── Courbes nDCG par k ────────────────────────────────────────
fig, ax = plt.subplots(figsize=(11, 6))
for (nom, vals), col in zip(ndcg_results.items(), colors_m):
    ax.plot(range(len(k_values_dcg)), vals, color=col, lw=2.5,
            marker='o', markersize=8, label=nom)

ax.set_xticks(range(len(k_values_dcg)))
ax.set_xticklabels([f'@{k}' if k < len(y_test) else '@all' for k in k_values_dcg],
                    fontsize=11)
ax.set_xlabel('Seuil k (position dans le classement)', fontsize=12)
ax.set_ylabel('nDCG@k', fontsize=12)
ax.set_title('nDCG@k — Qualite du classement par modele\n'
             'Un nDCG proche de 1 indique que les benignes sont bien classes en tete',
             fontsize=12, fontweight='bold')
ax.set_ylim(0.8, 1.05)
ax.legend(fontsize=10); ax.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

## Interpretation methodologique

Un modele de classification medicale est evalue a deux niveaux :

bonnes predictions ? On mesure accuracy, F1, AUC.

**Niveau 2 — Fiabilite structurelle** : est-ce que la structure des
probabilites predites ressemble a la realite ? On mesure Dice, SSIM, nDCG.
Un modele peut avoir une bonne accuracy mais une mauvaise calibration (SSIM
faible), ce qui le rend dangereux en pratique clinique.


```python
# ============================================================
# ETAPE 14 — DASHBOARD FINAL : SYNTHESE VISUELLE COMPLETE
# ============================================================
#
# Ce dashboard presente en un seul visuel l'ensemble des
# resultats obtenus. Chaque sous-graphique repond a une
# question precise sur la qualite des modeles.
# ============================================================

# Collecte de toutes les metriques pour les 4 modeles
from sklearn.metrics import (
    log_loss, cohen_kappa_score, matthews_corrcoef, roc_auc_score
)

def toutes_metriques(y_true, y_pred, y_proba, nom):
    cm_           = confusion_matrix(y_true, y_pred)
    tn_, fp_, fn_, tp_ = cm_.ravel()
    acc_  = accuracy_score(y_true, y_pred)
    prec_ = tp_ / (tp_ + fp_) if (tp_+fp_) > 0 else 0
    rec_  = tp_ / (tp_ + fn_) if (tp_+fn_) > 0 else 0
    spec_ = tn_ / (tn_ + fp_) if (tn_+fp_) > 0 else 0
    f1_   = 2*prec_*rec_ / (prec_+rec_) if (prec_+rec_) > 0 else 0
    fpr_  = fp_ / (fp_ + tn_) if (fp_+tn_) > 0 else 0
    return {
        'Modele'     : nom,
        'Accuracy'   : acc_,
        'F1'         : f1_,
        'AUC'        : roc_auc_score(y_true, y_proba),
        'Recall'     : rec_,
        'Specificite': spec_,
        'FPR'        : fpr_,
        'Log Loss'   : log_loss(y_true, y_proba),
        'Kappa'      : cohen_kappa_score(y_true, y_pred),
        'MCC'        : matthews_corrcoef(y_true, y_pred),
        'FN'         : int(fn_),
        'FP'         : int(fp_),
        'Dice'       : dice_vals[nom],
        'ASD'        : asd_vals[nom],
        'SSIM'       : ssim_vals[nom],
        'nDCG'       : ndcg_results[nom][-1],
    }

records = [
    toutes_metriques(y_test, y_pred_nb,     y_proba_nb,  'Naive Bayes'),
    toutes_metriques(y_test, y_pred_knn_sk, y_proba_knn, f'KNN (k={k_optimal})'),
    toutes_metriques(y_test, y_pred_dt,     y_proba_dt,  f'DT (depth={best_depth})'), # Corrected here
    toutes_metriques(y_test, y_pred_rf,     y_proba_rf,  f'RF ({best_n} arbres)'), # Corrected here
]
df_all = pd.DataFrame(records).set_index('Modele')

noms_dash  = df_all.index.tolist()
noms_court = ['NB', 'KNN', 'DT', 'RF']
colors_4   = ['#2563EB', '#16A34A', '#F59E0B', '#C0392B']

# ── Dashboard 6 sous-graphiques ───────────────────────────────
fig = plt.figure(figsize=(20, 14))
fig.suptitle(
    'Dashboard Final — Evaluation complete des 4 modeles\n'
    'Dataset Wisconsin Breast Cancer (569 patients, 30 features)',
    fontsize=15, fontweight='bold', y=1.01
)
gs = gridspec.GridSpec(3, 3, figure=fig, hspace=0.55, wspace=0.38)

# ── Sous-graphique 1 : Accuracy, F1, AUC ─────────────────────
ax1 = fig.add_subplot(gs[0, 0])
x   = np.arange(len(noms_court))
w   = 0.25
ax1.bar(x - w, [df_all.loc[n,'Accuracy']*100 for n in noms_dash],
        w, color='#2563EB', alpha=0.85, label='Accuracy')
ax1.bar(x,     [df_all.loc[n,'F1']*100      for n in noms_dash],
        w, color='#16A34A', alpha=0.85, label='F1')
ax1.bar(x + w, [df_all.loc[n,'AUC']*100     for n in noms_dash],
        w, color='#F59E0B', alpha=0.85, label='AUC')
ax1.set_xticks(x); ax1.set_xticklabels(noms_court, fontsize=11)
ax1.set_ylim(80, 105); ax1.set_ylabel('%')
ax1.set_title('Accuracy / F1 / AUC', fontweight='bold')
ax1.legend(fontsize=8); ax1.grid(axis='y', alpha=0.2)

# ── Sous-graphique 2 : Faux Negatifs (critere clinique) ───────
ax2 = fig.add_subplot(gs[0, 1])
fn_vals = [df_all.loc[n,'FN'] for n in noms_dash]
bars2   = ax2.bar(noms_court, fn_vals, color=colors_4, alpha=0.85,
                  edgecolor='white', width=0.55)
ax2.set_title('Faux Negatifs\n(cancers non detectes — MINIMISER)',
              fontweight='bold')
ax2.set_ylabel('Nombre de FN')
ax2.grid(axis='y', alpha=0.2)
for bar, v in zip(bars2, fn_vals):
    ax2.text(bar.get_x()+bar.get_width()/2, v+0.05, str(v),
             ha='center', va='bottom', fontsize=14, fontweight='bold',
             color='#DC2626' if v > 0 else '#16A34A')

# ── Sous-graphique 3 : Kappa et MCC ──────────────────────────
ax3 = fig.add_subplot(gs[0, 2])
kappa_v = [df_all.loc[n,'Kappa'] for n in noms_dash]
mcc_v   = [df_all.loc[n,'MCC']   for n in noms_dash]
ax3.bar(x - w/2, kappa_v, w, color='#7C3AED', alpha=0.85, label="Cohen Kappa")
ax3.bar(x + w/2, mcc_v,   w, color='#0891B2', alpha=0.85, label='MCC')
ax3.set_xticks(x); ax3.set_xticklabels(noms_court, fontsize=11)
ax3.set_title("Cohen's Kappa et MCC\n(accord au-dela du hasard)", fontweight='bold')
ax3.legend(fontsize=9); ax3.grid(axis='y', alpha=0.2)
for i, (k_v, m_v) in enumerate(zip(kappa_v, mcc_v)):
    ax3.text(i - w/2, k_v + 0.01, f'{k_v:.3f}', ha='center', fontsize=8, fontweight='bold')
    ax3.text(i + w/2, m_v + 0.01, f'{m_v:.3f}', ha='center', fontsize=8, fontweight='bold')

# ── Sous-graphique 4 : Courbes ROC superposees ────────────────
ax4 = fig.add_subplot(gs[1, 0])
all_pr = {
    'Naive Bayes'          : y_proba_nb,
    f'KNN (k={k_optimal})' : y_proba_knn,
    f'DT (depth={best_depth})' : y_proba_dt, # Corrected here
    f'RF ({best_n} arbres)'  : y_proba_rf, # Corrected here
}
for (nom, prob), col in zip(all_pr.items(), colors_4):
    fpr_r, tpr_r, _ = roc_curve(y_test, prob)
    auc_r            = roc_auc_score(y_test, prob)
    ax4.plot(fpr_r, tpr_r, color=col, lw=2,
             label=f'{nom[:12]}  {auc_r:.3f}')
ax4.plot([0,1],[0,1],'--',color='gray',lw=1.2)
ax4.set_xlabel('FPR'); ax4.set_ylabel('TPR')
ax4.set_title('Courbes ROC', fontweight='bold')
ax4.legend(fontsize=8, loc='lower right'); ax4.grid(alpha=0.2)

# ── Sous-graphique 5 : Dice, SSIM ────────────────────────────
ax5 = fig.add_subplot(gs[1, 1])
dice_v = [dice_vals[n] for n in noms_dash]
ssim_v = [ssim_vals[n] for n in noms_dash]
ax5.bar(x - w/2, dice_v, w, color='#2563EB', alpha=0.85, label='Dice')
ax5.bar(x + w/2, ssim_v, w, color='#16A34A', alpha=0.85, label='SSIM')
ax5.set_xticks(x); ax5.set_xticklabels(noms_court, fontsize=11)
ax5.set_title('Dice Score et SSIM\n(similarite structurelle)', fontweight='bold')
ax5.legend(fontsize=9); ax5.grid(axis='y', alpha=0.2)
for i, (d_v, s_v) in enumerate(zip(dice_v, ssim_v)):
    ax5.text(i-w/2, d_v+0.002, f'{d_v:.3f}', ha='center', fontsize=8, fontweight='bold')
    ax5.text(i+w/2, s_v+0.002, f'{s_v:.3f}', ha='center', fontsize=8, fontweight='bold')

# ── Sous-graphique 6 : Clustering ────────────────────────────
ax6 = fig.add_subplot(gs[1, 2])
clust_noms = ['K-Means', 'Hierarchique']
sil_vals_c = [sil_final,   sil_h_final]
db_vals_c  = [db_final,    db_h_final]
xc = np.arange(2)
ax6.bar(xc - 0.2, sil_vals_c, 0.35, color='#7C3AED', alpha=0.85, label='Silhouette')
ax6.bar(xc + 0.2, db_vals_c,  0.35, color='#F59E0B', alpha=0.85, label='Davies-Bouldin')
ax6.set_xticks(xc); ax6.set_xticklabels(clust_noms, fontsize=11)
ax6.set_title('Qualite du Clustering\nSilhouette (haut=bon) / DB (bas=bon)',
              fontweight='bold')
ax6.legend(fontsize=9); ax6.grid(axis='y', alpha=0.2)
for i, (s_v, d_v) in enumerate(zip(sil_vals_c, db_vals_c)):
    ax6.text(i-0.2, s_v+0.01, f'{s_v:.3f}', ha='center', fontsize=9, fontweight='bold')
    ax6.text(i+0.2, d_v+0.01, f'{d_v:.3f}', ha='center', fontsize=9, fontweight='bold')

# ── Sous-graphique 7 : Radar Chart ───────────────────────────
ax7 = fig.add_subplot(gs[2, :2], polar=True)
radar_m = ['Accuracy','F1','AUC','Recall','Specificite','Kappa','MCC','Dice','SSIM','nDCG']
radar_l = ['Acc','F1','AUC','Recall','Spec.','Kappa','MCC','Dice','SSIM','nDCG']
N_r     = len(radar_m)
angles  = np.linspace(0, 2*np.pi, N_r, endpoint=False).tolist()
angles += angles[:1]

for nom, col in zip(noms_dash, colors_4):
    vals_r  = [df_all.loc[nom, m] for m in radar_m]
    vals_r += vals_r[:1]
    ax7.plot(angles, vals_r, color=col, lw=2.0, label=nom)
    ax7.fill(angles, vals_r, color=col, alpha=0.04)

ax7.set_xticks(angles[:-1])
ax7.set_xticklabels(radar_l, size=11)
ax7.set_ylim(0, 1)
ax7.set_title('Radar Chart — Vue multi-criteres\n'
              'Plus le polygone est large, meilleur est le modele',
              fontweight='bold', pad=20)
ax7.legend(loc='upper right', bbox_to_anchor=(1.45, 1.1), fontsize=9)
ax7.grid(color='gray', alpha=0.3)

# ── Sous-graphique 8 : Tableau recapitulatif ──────────────────
ax8 = fig.add_subplot(gs[2, 2])
ax8.axis('off')

recap_cols = ['Acc%','F1%','AUC','FN','Kappa','MCC','Dice','SSIM']
table_data = []
for nom in noms_dash:
    r = df_all.loc[nom]
    table_data.append([
        f'{r["Accuracy"]*100:.1f}',
        f'{r["F1"]*100:.1f}',
        f'{r["AUC"]:.3f}',
        str(int(r["FN"])),
        f'{r["Kappa"]:.3f}',
        f'{r["MCC"]:.3f}',
        f'{r["Dice"]:.3f}',
        f'{r["SSIM"]:.3f}',
    ])

tbl = ax8.table(
    cellText    = table_data,
    rowLabels   = noms_court,
    colLabels   = recap_cols,
    cellLoc     = 'center',
    loc         = 'center',
)
tbl.auto_set_font_size(False)
tbl.set_fontsize(9)
tbl.scale(1.15, 1.5)

# Mise en evidence du meilleur par colonne
best_idx = {
    0: int(np.argmax([df_all.loc[n,'Accuracy']   for n in noms_dash])),
    1: int(np.argmax([df_all.loc[n,'F1']         for n in noms_dash])),
    2: int(np.argmax([df_all.loc[n,'AUC']        for n in noms_dash])),
    3: int(np.argmin([df_all.loc[n,'FN']         for n in noms_dash])),
    4: int(np.argmax([df_all.loc[n,'Kappa']      for n in noms_dash])),
    5: int(np.argmax([df_all.loc[n,'MCC']        for n in noms_dash])),
    6: int(np.argmax([df_all.loc[n,'Dice']       for n in noms_dash])),
    7: int(np.argmax([df_all.loc[n,'SSIM']       for n in noms_dash])),
}
for col_idx, row_idx in best_idx.items():
    tbl[(row_idx+1, col_idx)].set_facecolor('#DCFCE7')

ax8.set_title('Recapitulatif (vert = meilleur)', fontweight='bold', pad=12)

plt.tight_layout()
plt.savefig('dashboard_final.png', dpi=150, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
print('Dashboard sauvegarde : dashboard_final.png')
```
